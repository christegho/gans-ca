```python
from selenium import webdriver
import urllib
scroll_pause_time = 2
scroll_n = 4000
import time

"""
https://stackoverflow.com/questions/28928068/scroll-down-to-bottom-of-infinite-page-with-phantomjs-in-python/28928684#28928684
https://medium.com/@erika_dike/how-to-download-100-pictures-from-a-site-with-selenium-e23b7ecacb85
https://stackoverflow.com/questions/45781388/how-to-download-all-images-from-webpage-using-python
"""
class WebPageDownloader:
    def __init__(self,url):
        self.url=url
        self.uri = []
        self.folder = './'
        self.im_i = 0
    def download_images(self):
        self.driver=webdriver.Firefox()
        self.driver.get(self.url)
    
        for i in range(scroll_n):  
            srcs = []
            if i % 10 == 0:
                print(i)
            self.driver.execute_script("window.scrollTo("+ str(8000*i) + ", " + str(8000*(i+1)) + ")") 
            time.sleep(scroll_pause_time)
        self.r=self.driver.find_elements_by_tag_name('img')
        for v in self.r:
            src = v.get_attribute("src")
            if not src:
                src = v.get_attribute("data-src")
            if src not in self.uri:
                self.uri.append(src)
                srcs.append(src)
                pos = len(src) - src[::-1].index('/')
#                 print src[pos:]
                self.im_i += 1
                self.g=urllib.urlretrieve(src, "/".join([self.folder, str(self.im_i)+'.jpg']))
            if self.im_i%100 == 0:
                print(self.im_i)
            

if __name__=='__main__':
    PAGEDOWNLOAD=WebPageDownloader("http://www.contemporaryartdaily.com/")
    PAGEDOWNLOAD.download_images()
```
